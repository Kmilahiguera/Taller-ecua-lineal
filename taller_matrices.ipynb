```python
import numpy as np
import matplotlib.pyplot as plt
```

Se define las ecuaciones;

- 6x+3y-3z= 52
- 23x+5y+6z = -25
- -4x+4y-17z = -9

Se pasan a una matriz, y la matriz resulatdo


```python
a= np.array([[6, 3, -3], [23, 5, 6],[-4, 4, -17]])
b= np.array([52, -25, -9])

a.shape
```

Resolver las ecuaciones lineales

```python
Solve= np.linalg.solve(a, b)
```

Mostrar las soluciones

```python
print("la solcuión es :")
print("X=", Solve[0])
print("y=", Solve[1])
print("z=", Solve[2])
```

Graficar soluciones:


1. Crear figura en la cual se pueda graficar

```python
fig= plt.figure()
```

2. Agregar un grafico deentro de esta figura (determinar ejes)

```python
ax= fig.add_subplot(111, projection='3d')
```

3. determianr el tamaño de los ejes y crear la grilla de coordenadas

```python
x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)
x,y=np.meshgrid(x,y)
```

Calular z con la ecuacion implicita de los planos

```python
  Z1 = (-a[0,0]*x -a[0,1]*y +b[0])/ a[0,2]
  Z2 = (-a[1,0]*x - a[1,1]*y + b[1]) / a[1,2]
  Z3 = (-a[2,0]*x - a[2,1]*y + b[2]) / a[2,2]
```

Dibujar laa  superficie en tridimensional en cada eje con sus respectivos  parametroa

```python
ax.plot_surface(x, y, Z1, alpha=0.5)
ax.plot_surface(x, y, Z2, alpha=0.5)
ax.plot_surface(x, y, Z3, alpha=0.5)
```

Crear las etiquedas cada cada eje

```python
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel('Z')
```

Tomando todos los pasos se procedea mostrar el grafico

```python
fig= plt.figure()
ax= fig.add_subplot(111, projection='3d')

x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)
x,y=np.meshgrid(x,y)


Z1 = (-a[0,0]*x -a[0,1]*y +b[0])/ a[0,2]
Z2 = (-a[1,0]*x - a[1,1]*y + b[1]) / a[1,2]
Z3 = (-a[2,0]*x - a[2,1]*y + b[2]) / a[2,2]

ax.plot_surface(x, y, Z1, alpha=0.1)
ax.plot_surface(x, y, Z2, alpha=0.1)
ax.plot_surface(x, y, Z3, alpha=0.1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel('Z')

plt.show()
```
